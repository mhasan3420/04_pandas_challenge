# Heroes of Pymoli

```python
import pandas as pd
```

```python
# The path to our CSV file
file_to_load = "02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read our data into pandas
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(2)
```

# Player Count


```python
total_players = len(purchase_data["SN"].value_counts())
total_players_df = pd.DataFrame([total_players], columns = ["Total Players"])
total_players_df
```

# Purchasing Analysis (Total)


```python
#Basic Calculations
item_count = len(purchase_data["Item ID"].unique())
total_purchase_value = purchase_data["Price"].sum()
purchase_count = purchase_data["Price"].count()
average_item_price = purchase_data["Price"].mean()

# Create DataFrame 
purchasing_analysis = pd.DataFrame({"Number of Unique Items": item_count,
                               "Total Revenue": [total_purchase_value],
                               "Number of Purchases": [purchase_count],
                               "Average Price": [average_item_price]})
# Data Munging
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${:.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:}".format)

summary_table = purchasing_analysis.loc[:, ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

summary_table
```

#  Gender Demographics


```python
#Basic Calculations
gender_list = purchase_data.loc[:,["Gender", "SN"]]
gender_list = gender_list.drop_duplicates()
gender_total = gender_list["Gender"].value_counts()
gender_percent = gender_total / total_players *100
# Create DataFrame 
gender_demographics = pd.DataFrame({"Total Count": gender_total,
                                    "Percentage of Players": gender_percent})
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}".format)
gender_demographics
```

#  Purchasing Analysis (Gender)

```python
#Basic Calculations
gender_analysis_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
gender_average_price = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Value")
gender_total_purchase = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

average_tp_perperson = gender_total_purchase / gender_total
# Create DataFrame 
gender_purchasing_analysis = pd.DataFrame({"Purchase Count": gender_analysis_total,
                                           "Average Purchase Price": gender_average_price,
                                           "Total Purchase Value": gender_total_purchase,
                                           "Avg Total Purchase per Person": average_tp_perperson})
#Data Formating 
gender_purchasing_analysis["Average Purchase Price"] = gender_purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_purchasing_analysis["Total Purchase Value"] = gender_purchasing_analysis["Total Purchase Value"].map("${:.2f}".format)
gender_purchasing_analysis["Avg Total Purchase per Person"] = gender_purchasing_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

summary_table = gender_purchasing_analysis.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
summary_table 
```

#  Top Spenders

```python
#Basic Calculations
sn_list = purchase_data.loc[:,["SN"]]
sn_list = sn_list.drop_duplicates()
sn_existing_list = sn_list["SN"].value_counts()

sn_analysis_total = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
sn_average_price = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
sn_total_purchase = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
#summary data frame
sn_purchasing_analysis = pd.DataFrame({"Purchase Count": sn_analysis_total,
                              "Average Purchase Price": sn_average_price,
                              "Total Purchase Value": sn_total_purchase})
#sorting and displayed 
sn_purchasing_analysis = sn_purchasing_analysis.sort_values("Total Purchase Value", ascending=False)
#formating data 
sn_purchasing_analysis["Average Purchase Price"]=sn_purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
sn_purchasing_analysis["Total Purchase Value"]=sn_purchasing_analysis["Total Purchase Value"].map("${:.2f}".format)
sn_purchasing_analysis.head()
```

#  Most Popular Items

```python
#Retrive and groupby 
item_id = purchase_data.groupby("Item ID").sum().reset_index()
item_name_sum = purchase_data.groupby("Item Name").sum().reset_index()
item_name_count = purchase_data.groupby("Item Name").count().reset_index()
# merge items 
item_id_sum = pd.merge(item_id, item_name_sum, on="Price")
item_id_count = pd.merge(item_name_count, item_id_sum, on="Item Name")
item_id_count["Gender"] = (item_id_count["Price_y"]/item_id_count["Item ID"]).round(2)
#summary data frame
item_id_count_rename = item_id_count.rename(columns={"Age": "Purchase Count", "Gender": "Item Price", "Item ID": "null", "Price_y": "Total Purchase Value", "Item ID_y": "Item ID"})
clean_item = item_id_count_rename[["Item ID", "Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]]
popular_item = clean_item.set_index(["Item ID", "Item Name"])

#Sort the total purchase value column in descending order
popular_item_total = popular_item.sort_values("Purchase Count", ascending=False).head(5)
popular_item_total.style.format({"Item Price": "${:.2f}", "Total Purchase Value": "${:.2f}"})
```
